# Strain-specific iModulons
Two iModulons- currently labeled `Genomic Island 3-1` and `Genomic Island 1-2`- capture much of the USA300 vs non-USA300 (NCTC) differences. Rename them into `USA300 iModulon-1` and `USA300 iModulon-2`.

### USA300 iModulon-1
Plot the iModulon gene content and highlight key genes e.g. pathogenic islands, SCCMec, prophages etc.

```python
imod_u1 = 'USA300 iModulon-1'
ica_data.change_threshold('USA300 iModulon-1', 0.0281)
imod_genes = ica_data.view_imodulon(imod_u1)
non_imod = ica_data.gene_table.index.difference(imod_genes.index)
imod =  ica_data.gene_table.loc[imod_genes.index].sort_values('start')
# acme = imod[10:45]
# sccmec = imod[:10]
sapi5 = imod[imod.operon.isin(['Op564', 'Op565', 'Op566', 'Op567', 'Op568',
                               'Op569', 'Op570', 'Op571'])]
is_seq = imod[imod.operon.isin(['Op416','Op417', 'Op418',
                                'Op419', 'Op420', 'Op421', 'Op422', 'Op423'])]
phi_pvl = imod[imod.operon.str.match(r'Op9[0-9][0-9]')]
plasmid = imod.loc[['USA300HOU_RS15640', 'USA300HOU_RS14735']]
sccmec = imod[(imod_genes.start >= 34513) & (imod_genes.end < 57875)]
imod = imod.drop(sapi5.index | is_seq.index | phi_pvl.index |
                 plasmid.index | sccmec.index)

clr = sns.palettes.color_palette('Set2')

fig, ax = plt.subplots()
ax.scatter(ica_data.gene_table.loc[non_imod, 'start'], ica_data.M.loc[non_imod, imod_u1],
          color='lightgray')
ax.scatter(sccmec['start'], ica_data.M.loc[sccmec.index, imod_u1],
           label='SCCMecIVa', color=clr[1])
# ax.scatter(acme['start'], ica_data.M.loc[acme.index, imod_u1],
#            label='ACME', color=clr[0])
ax.scatter(is_seq['start'], ica_data.M.loc[is_seq.index, imod_u1],
           label='13kb IS', color=clr[2])
ax.scatter(sapi5['start'], ica_data.M.loc[sapi5.index, imod_u1],
           label='SaPI5', color=clr[4])
ax.scatter(phi_pvl['start'], ica_data.M.loc[phi_pvl.index, imod_u1],
           label='PhiPVL', color=clr[5])

ax.scatter(imod['start'], ica_data.M.loc[imod.index, imod_u1],
           label='None', color='#1b9e77')

ax.scatter(plasmid['start'], ica_data.M.loc[plasmid.index, imod_u1],
           label='pUSA300HOUMR', color='#d95f02')

# ax.scatter(ica_data.gene_table.loc[imod_genes.index, 'start'],
#            ica_data.M.loc[imod_genes.index, imod_u1])

ax.hlines(0.03, -100000, 3e6, color='gray', linestyle='--')
ax.hlines(-0.03, -100000, 3e6, color='gray', linestyle='--')
ax.set_xlim(-100000, 3e6)

ax.tick_params(which='both', axis='both', direction='in', labelsize=14)
ax.tick_params(which='major', axis='both', direction='in', length=6)

ax.set_xlabel('Gene Start Position (Mbp)', fontname='Arial', fontsize=16)
ax.set_ylabel('Gene Weight', fontname='Arial', fontsize=16)
ax.xaxis.set_minor_locator(MultipleLocator(2e4))
ax.yaxis.set_minor_locator(MultipleLocator(0.01))


plt.legend(frameon=False, title='Mobile Elements')
# ax.text(119127, 0.21, '$mecA$')
# ax.text(119127, ica_data.M.loc['USA300HOU_RS00325', imod_u1], '$arcA$')
# ax.text(119127, ica_data.M.loc['USA300HOU_RS00325', imod_u1], '$arcA$')

ax.text(ica_data.gene_table.loc['USA300HOU_RS07645', 'start'] + 80000,
       ica_data.M.loc['USA300HOU_RS07645', imod_u1],
       '$lukSF$')
ax.text(ica_data.gene_table.loc['USA300HOU_RS04430', 'start'] + 80000,
       ica_data.M.loc['USA300HOU_RS04430', imod_u1],
       '$sek$')

ax.text(ica_data.gene_table.loc['USA300HOU_RS04425', 'start'] + 80000,
       ica_data.M.loc['USA300HOU_RS04425', imod_u1],
       '$seq$')

ax.text(ica_data.gene_table.loc['USA300HOU_RS01770', 'start'] - 200000,
        ica_data.M.loc['USA300HOU_RS01770', imod_u1] - 0.012,
        '$mepR$')
ax.text(ica_data.gene_table.loc['USA300HOU_RS02945', 'start'],
        ica_data.M.loc['USA300HOU_RS02945', imod_u1] + 0.005,
        '$sdrC$')
ax.text(ica_data.gene_table.loc['USA300HOU_RS03610', 'start'] + 40000,
        ica_data.M.loc['USA300HOU_RS03610', imod_u1],
        '$sarX$')
ax.text(ica_data.gene_table.loc['USA300HOU_RS09190', 'start'],
        ica_data.M.loc['USA300HOU_RS09190', imod_u1] + 0.01,
        '$isdH$')
ax.text(ica_data.gene_table.loc['USA300HOU_RS09715', 'start'],
        ica_data.M.loc['USA300HOU_RS09715', imod_u1] - 0.015,
        '$lukD$')
ax.text(ica_data.gene_table.loc['USA300HOU_RS14330', 'start'] - 40000,
        ica_data.M.loc['USA300HOU_RS14330', imod_u1] - 0.01,
        '$argR1$')



plt.savefig('../../figures/Strain_imod_Gene_weight_main.svg')
```

# Load DBGWAS k-mer data

```python
kmer_md = pd.read_csv('../data/dbgwas/dbgwas_postprocess/kmer_md.csv', index_col=0)
kmer_md.TCH1516_feats = kmer_md.TCH1516_feats.fillna('')

all_nodes = pd.read_csv('../data/dbgwas/dbgwas_output/textualOutput/all_comps_nodes_info.tsv',
                        sep='\t')
all_nodes.reset_index(inplace=True)
all_nodes.drop('Annotations(sep=~~~)', inplace=True, axis=1)
all_nodes.columns = ['CompId', 'NodeId', 'AlleleFreq', 'Pheno0Count',
       'Pheno0TotalCount', 'Pheno1Count', 'Pheno1TotalCount', 'NACount',
       'NATotalCount', 'Significant?', 'p-value', 'q-Value', 'EstEffect',
       'WaldStat', 'Sequence', 'SequenceLength', 'Annotations(sep=~~~)']
all_nodes.loc[:, 'Annotations(sep=~~~)'] = all_nodes['Annotations(sep=~~~)'].fillna("No Annotation")

sig_nodes = all_nodes[all_nodes['Significant?'] == 'Yes'].reset_index()
```

```python
sig_genes = set(itertools.chain(*[i.split(';') for i in kmer_md.TCH1516_feats.values])) 
sig_genes = [i.replace('(intergenic)', '') for i in sig_genes][1:] # 1: takes out the empty name
print(f'Number of unique genes associated with significant kmer: {len(set(sig_genes))}')
print(f'Number of unique components: {len(sig_nodes.CompId.unique())}')
```

```python
# number of genes in strain-specific iModulons and enriched in DBGWAS
sg_imod1 = ica_data.view_imodulon('USA300 iModulon-1').index.intersection(sig_genes)
sg_imod2 = ica_data.view_imodulon('USA300 iModulon-2').index.intersection(sig_genes)
print(len(sg_imod1), len(sg_imod2))
```

```python
# all genes with sig kmers + member of sig iModulons
sg_imods_combined = set(sg_imod1).union(sg_imod2)
```

```python
ica_data.gene_table.loc[sg_imods_combined].to_csv('../tables/iModanddbgwasintersection.csv')
```

## Get conserved genes between Non-USA300 (NCTC Reference) and USA300 with Bi-Directional BLAST Hits (BBH)

```python
# create protein fasta file and BLAST db
compare.make_prots('../data/genomes/TCH1516.gb', '../data/genomes/TCH1516.faa')
compare.make_prot_db('../data/genomes/TCH1516.faa')


compare.make_prots('../data/genomes/NCTC8325.gb', '../data/genomes/NCTC8325.faa')
compare.make_prot_db('../data/genomes/NCTC8325.faa')
```

```python
bbh = compare.get_bbh('../data/genomes/TCH1516.faa',
                     '../data/genomes/NCTC8325.faa',
                     savefiles=True,
                     outdir='../data/genomes/')
```

```python
bbh = bbh.rename(columns={'gene': 'TCH1516', 'subject': 'NCTC8325'})
bbh = bbh[bbh.BBH == '<=>']
print(f'found {len(bbh)} shared genes between NCTC8325(non-USA300) and TCH1516(USA300)')
```

```python
# intersection of genes associated with USA300 & found in strain-specific iModulon and conserved in TCH1516 and NCTC8325
intersections = ica_data.gene_table.loc[set(bbh.TCH1516.values).intersection(sg_imods_combined)]
intersections.to_csv('../tables/iModanddbgwasandbbhintersection.csv')
```
